# Logistic Regression using Breast Cancer datasets

we uild up the tools to build a logistic Regression model  for a classification datasets, we'll introduce a new datset.

In the breast cancer datset, each datapoints has measurements from an image of a breast mass and whether or not it's cancerou. The goal will be to use these measurements to predict if the mass is cencerious.

This dataset is build right into scikit-learn so we won't need to read a csv.

Let's start by loading the dataset and taking a peak at the data and how it's formatted.

```
import time
import pandas as pd
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer()
```

The object returned (which we stored in the cancer_data variable) is an object similar to a python dictionary. we can see the available keys with the kes method.

```
print(cancer_data.keys())
print(cancer_data['DESCR'])
```

```
from  sklearn.linear_model import LogisticRegression
```

```
df=pd.DataFrame(cancer_data['data'], columns=cancer_data['feature_names'])
df['target'] = cancer_data['target']
```

```
X=df[cancer_data.feature_names].values
y=df['target'].values
```

```
df.head()
```

```
model = LogisticRegression(solver='liblinear')
model.fit(X,y)
```

```
print('predction for datapoint0:',model.predict([X[0]]))
print(model.score(X,y))
```
